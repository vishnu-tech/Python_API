```python
import requests
api_key = "726b2e76bc9d66c75617e8e61bd317d1" 
api_url = "https://api.themoviedb.org/3" 
params = {'query':"game" , 'api_key':api_key} 
header = {'Accept': 'application/json'} 
response = requests.get(api_url + "/search/tv/", headers = header, params = params) 
data = response.json() 
results = data.get('results') 

for result in results: 
    print(result)
```

```python
# https://api.nasa.gov/mars-photos/api/v1/rovers/curiosity/photos?sol=1000&api_key=DEMO_KEY

import requests 
api_key = "DEMO_KEY" 
api_url = "https://api.nasa.gov" 

params = {'sol':"1000" , 'api_key':api_key}

header = {'Accept': 'application/json'} 

response = requests.get(api_url + "/mars-photos/api/v1/rovers/curiosity/photos/", headers = header, params = params) 
print(response.status_code)

#if we remove API key then error by removing params
```

```python
data = response.json() 

photos_data = data.get('photos')

for photo_data in photos_data:
    print(photo_data['camera']['full_name'])
```

```python
from skimage import io

data = response.json() 

photos_data = data.get('photos')

for photo_data in photos_data[0:1]:
    img_url = photo_data['img_src']
    img = io.imread(img_url)
    io.imshow(img)
    print(photo_data['camera']['full_name'])
```
