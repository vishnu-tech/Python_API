```python
import requests

response = requests.get('https://api.coindesk.com/v1/bpi/currentprice.json')

data = response.json()
print(data)
```

```python
print("Time: " + str(data['time']['updated']))
print("Price 1 BTC : " + str(data['bpi']['USD']['rate_float']))
```

```python
# https://api.nasa.gov/mars-photos/api/v1/rovers/curiosity/photos?sol=1000&api_key=DEMO_KEY

import requests 
api_key = "DEMO_KEY" 
api_url = "https://api.nasa.gov" 

params = {'sol':"1000" , 'api_key':api_key}

header = {'Accept': 'application/json'} 

response = requests.get(api_url + "/mars-photos/api/v1/rovers/curiosity/photos/", headers = header, params = params) 
print(response.status_code)

#if we remove API key then error by removing params
```

```python
data = response.json()

photos_data = data.get('photos')

for photo_data in photos_data:
    print(photo_data['camera']['full_name'])
```
